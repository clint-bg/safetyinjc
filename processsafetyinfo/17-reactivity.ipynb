# Chemical Reactivity

```{seealso} Learning Outcomes
- Use Gibbs free energy and other thermodynamic properties to estimate the stability and potential energy release of chemicals.
- Understand chemical compatibility and the risks of accidental mixing of incompatible substances.
- Interpret data from thermal stability tests (e.g., DSC, ARC) to determine safe processing limits.
```

```{note} Reading
- Foundations of Spiritual and Physical Safety: with Chemical Processes; Chapter 6, Sec. 2 to Sec. 2.1
```
