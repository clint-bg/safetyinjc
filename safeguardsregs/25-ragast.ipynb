Just like RAGAGEP (Recognized and Generally Accepted Good Engineering Practices) that help us work safely and effectively with chemical processes, there are spiritually based guidelines and practices that help us to prepare for life after death. Those guidelines and practices are recorded in scripture.

## Accepted Scriptural Texts

There are many religions in the world that have greatly benefited humanity.

### Bible

Good ethical practices highlighted in the Bible are many and most notably include the teachings of Jesus Christ on the Sermon on the Mount:
- Love your enemies
- Be merciful
- Be humble
- Forgive
- Hunger and thirst after righteousness
- Be a peacemaker
- Be a light to those around you
- Avoid lust and adultery
- Avoid anger
- Avoid divorce
- Be honest 
- Be perfect like Heavenly Father

### Book of Mormon

Joseph Smith\index{joseph smith}\index{book of mormon} was called by God in a miraculous way in 1820. He saw and spoke with God the Father and Jesus Christ\index{Jesus Christ}, two distinct beings with bodies of flesh and bone. Joseph Smith translated the Book of Mormon from golden plates that had writings from ancient prophets in the Americas. Good ethical practices are ever present in the Book of Mormon: Another Testament of Jesus Christ, most notably during the visit of Jesus to the American continent after his death in Jerusalem. He taught the Sermon on the Mount again and witnessed of his sacrifice as God to make it possible for us to return to live with our Heavenly Father. His teaching as recorded in 3 Nephi include:

 - Calling and giving authority to baptize to Nephi and others
- The baptismal prayer and ordinance instruction
- Invitation for all to be baptized (make a covenant with God)
- The Sermon on the Mount
- Institution of the sacrament to remember Him regularly


I love the words in the Book of Mormon. Another of my favorite parts is the address of King Benjamin as recorded in the Book the Mosiah Chapters 2-4. A few of my favorite versus are:   

- ...believe in God; believe that he is and that he created all things, both in heaven and in earth...
- ...remember, and always retain in remembrance, the greatness of God, and your own nothingness, and his goodness and long-suffering towards you...
- ...there shall be no other name given nor any other way nor means whereby salvation can come unto the children of men, only in and through the name of Christ, the Lord Omnipotent...
- ...the Lord God hath sent his holy prophets among all the children of men, to declare these things...
- ...the natural man is an enemy to God unless he yields to the enticing of the Holy Spirit and becometh a saint though the atonement of Christ the Lord...
- ...when ye are in the service of your fellow beings ye are only in the service of your God...
