# On Becoming

```{seealso} Learning Outcomes
- Understand the concept of "becoming" as a long-term development of character rather than just compliance with rules.
- Align professional engineering decisions with the attributes of Jesus Christ, such as integrity and selfless service.
- Evaluate how daily safety and ethical choices contribute to one's eternal potential and relationship with God.
```

```{note} Reading
- Foundations of Spiritual and Physical Safety: with Chemical Processes; Chapter 2, Sections 3 and 7.
```
