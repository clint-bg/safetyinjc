# FMEA Case Study

```{seealso} Learning Outcomes
- Perform a Failure Mode and Effects Analysis (FMEA) by itemizing failure modes for specific equipment components.
- Analyze the consequences of each failure mode on the overall system and personnel safety.
- Assign risk rankings to failure modes and propose specific engineering or administrative controls to reduce risk.
```

```{note} Reading
- Read about an FMEA example on the CSB website.
```
