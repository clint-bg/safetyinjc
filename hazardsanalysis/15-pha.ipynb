# DHA and PHA

```{seealso} Learning Outcomes
- Differentiate between a Process Hazards Analysis (PHA) required by PSM and a Design Hazards Analysis (DHA).
- Understand the regulatory requirements for PHA under OSHA 1910.119, including team composition and revalidation schedules.
- Recognize the role of "Safety Action Requests" in tracking and implementing recommendations from these analyses.
```

```{note} Reading
- Foundations of Spiritual and Physical Safety: with Chemical Processes; Chapter 5, Sections 2.4-3
```
